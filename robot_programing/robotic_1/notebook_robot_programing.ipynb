  - Configuration space:
    - q = [q1, q2, q3, ..., qn] (a quantidade de q é igual a quantidade de juntas do robô / liberdade do robô)
      - onde cada qi é o valor de uma junta
  - joint space

- robot kinematics
  - Forward kinematics  (cinematica direta):
    - dado o valor das juntas, qual a posição do end-effector?
    - f(q) = x
      - q: joint space
      - x: task space
  - Inverse kinematics (cinematica inversa):
    - dado a posição do end-effector, qual o valor das juntas?
    - f^-1(x) = q
      - x: task space
      - q: joint space

- Inverse kinematics
  - Métodos analíticos:
    - Resolver as equações matemáticas que descrevem a cinemática do robô
    - Pode ter múltiplas soluções ou nenhuma solução
    - Exemplo: robô planar de 2 graus de liberdade
  - Métodos numéricos:
    - Usar algoritmos iterativos para encontrar a solução
    - Pode ser mais lento, mas é mais flexível
    - Exemplo: método de Newton-Raphson
  - Métodos baseados em aprendizado de máquina:
    - Usar redes neurais ou outros algoritmos de aprendizado para aprender a relação entre o espaço de tarefas e o espaço de juntas
    - Pode ser mais robusto a ruídos e incertezas
    - Exemplo: rede neural treinada com dados de simulação
  - Esse problema não tem uma receita de bolo
    - Cada estrutura mecânica do robô precisa de um método diferente para resolver a cinemática inversa
    - Calcular a distância entre o end-effector e o alvo usando o triângulo pitagôrico
  - Solução analítica para robô planar 2R:
    - x = l1*cos(q1) + l2*cos(q1 + q2)
    - y = l1*sin(q1) + l2*sin(q1 + q2)
    - Onde:
      - l1: comprimento do primeiro elo
      - l2: comprimento do segundo elo
      - q1: ângulo da primeira junta
      - q2: ângulo da segunda junta
    - Para encontrar q1 e q2 dados x e y, podemos usar as seguintes equações:
      - r = sqrt(x^2 + y^2)
      - cos(q2) = (r^2 - l1^2 - l2^2) / (2 * l1 * l2)
      - sin(q2) = sqrt(1 - cos^2(q2))
      - q2 = atan2(sin(q2), cos(q2))
      - k1 = l1 + l2 * cos(q2)
      - k2 = l2 * sin(q2)
      - q1 = atan2(y, x) - atan2(k2, k1)
    - Solução numérica usando o método de Newton-Raphson:
      - Começar com uma estimativa inicial para q1 e q2
      - Calcular a posição atual do end-effector usando a cinemática direta
      - Calcular o erro entre a posição atual e a posição desejada
      - Calcular a matriz Jacobiana do robô
      - Atualizar os valores de q1 e q2 usando a fórmula:
        - q_new = q_old + J_inv * error
      - Repetir até que o erro seja menor que um limiar pré-definido
