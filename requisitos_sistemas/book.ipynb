# Requisitos de Sistema

## 16-09-2025

### SysML vs UML
- SysML é uma extensão do UML, adaptada para engenharia de sistemas.
- UML é mais focado em desenvolvimento de software.
- SysML inclui diagramas adicionais para modelar requisitos, comportamento, estrutura e paramétricos.

### Diagramas SysML
- **Diagramas de Requisitos**: Capturam e organizam requisitos do sistema.
- **Diagramas de Blocos**: Modelam a estrutura do sistema.
- **Diagramas de Atividades**: Representam o fluxo de atividades e processos.
  - Fluxo de Controle
  - Fluxo de Objeto
- **Diagramas de Estados**: Mostram os estados e transições do sistema.
- **Diagramas de Sequência**: Ilustram a interação entre componentes ao longo do tempo.
- **Diagrama de Máquina de Estados**: Modela o comportamento dinâmico do sistema.
- **Diagrama de caso de uso**: Descreve as funcionalidades do sistema do ponto de vista do usuário.
- **Diagrama de Comportamento**: Representa o comportamento do sistema em resposta a eventos.
- **Diagrama de Pacotes**: Dividir em módulos/sub-sistemas -> componente composto por outros componentes.

### Diagramas mais utilizados
- Diagramas de Blocos
- Casos de Uso
- Máquina de estados - dependendo do contexto
